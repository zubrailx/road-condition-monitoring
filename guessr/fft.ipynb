```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a sample signal
sampling_frequency = 40  # Hz
window_size = 128
t = np.arange(window_size) / sampling_frequency
signal = np.sin(2 * np.pi * 10 * t)

# Compute the FFT
fft_result = np.fft.fft(signal)
frequencies = np.fft.fftfreq(window_size, 1/sampling_frequency)
print(frequencies)

# Plot the FFT result
plt.figure(figsize=(10, 4))
plt.plot(frequencies[:window_size//2], np.abs(fft_result[:window_size//2]))
plt.title('FFT Result')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid()
plt.show()
```
